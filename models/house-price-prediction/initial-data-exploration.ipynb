```python


# 1. Compute missing statistics
missing_stats = pd.DataFrame({
    "missing_count": dataset.isnull().sum(),
    "missing_%": dataset.isnull().mean() * 100
})

# 2. Sort by highest missing percentage
missing_stats = missing_stats.sort_values(by="missing_%", ascending=False)

# 3. Show only columns with missing values
missing_stats = missing_stats[missing_stats["missing_count"] > 0]

# 4. Display result
print(missing_stats)
```

```python
drop_cols = [
    "PoolQC", "MiscFeature", "Alley",
    "Fence", "MasVnrType", "FireplaceQu"
]

dataset = dataset.drop(columns=drop_cols)
```

```python
garage_cols = [
    "GarageType", "GarageQual", "GarageCond", "GarageFinish"
]

bsmt_cols = [
    "BsmtExposure", "BsmtFinType1", "BsmtFinType2",
    "BsmtCond", "BsmtQual"
]

dataset[garage_cols] = dataset[garage_cols].fillna("No_Garage")
dataset[bsmt_cols] = dataset[bsmt_cols].fillna("No_Basement")
```

```python
dataset["Electrical"] = dataset["Electrical"].fillna(
    dataset["Electrical"].mode()[0]
)
```

```python
dataset["GarageYrBlt"] = dataset["GarageYrBlt"].fillna(0)
```

```python
from sklearn.impute import KNNImputer

num_cols = dataset.select_dtypes(include="number").columns

imputer = KNNImputer(n_neighbors=5)
dataset[num_cols] = imputer.fit_transform(dataset[num_cols])
```

```python
#identify outliers
num_data = dataset.select_dtypes(include="number")
Q1 = num_data.quantile(0.25)
Q3 = num_data.quantile(0.75)
IQR = Q3 - Q1

outliers = (( num_data < (Q1 - 1.5 * IQR)) | ( num_data> (Q3 + 1.5 * IQR)))

print(outliers)
```

```python
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(dataset.select_dtypes(include="number")))
outliers = (z > 3)

print(outliers)
```

```python
for col in dataset.select_dtypes(include="number").columns:
    lower = dataset[col].quantile(0.01)
    upper = dataset[col].quantile(0.99)
    dataset[col] = dataset[col].clip(lower, upper)
```
